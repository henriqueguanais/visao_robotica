```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
def rotaciona_vizu(p, teta):
    matriz_rotacao = np.array([[np.cos(teta), -np.sin(teta)], [np.sin(teta), np.cos(teta)]])
    # plotar o ponto p e o ponto rotacionado    
    plt.plot(p[0], p[1], 'bo')
    p_rot = matriz_rotacao @ p
    plt.plot(p_rot[0], p_rot[1], 'ro')
    plt.plot([p_rot[0], 0], [p_rot[1], 0], 'r')
    plt.plot([p[0], 0], [p[1], 0], 'b')
    
    # area maxima do grafico
    max_val = np.max(np.abs([p, p_rot]))*1.2
    plt.xlim(-max_val, max_val)
    plt.ylim(-max_val, max_val)

    # destaca os eixos
    plt.axhline(0, color='black')
    plt.axvline(0, color='black')
    
    plt.grid()
    plt.show()
    print(f"O ponto: {p} foi rotacionado para: {p_rot}")

p = np.array([2, 4])
teta = np.pi/2

rotaciona_vizu(p, teta)
```

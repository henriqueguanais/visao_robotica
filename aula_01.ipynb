```python
# imagem com um quadrado vermelho
img_path = 'imagens/Lenna.png'
img = cv2.imread(img_path)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.show()

largura = img.shape[1]
altura = img.shape[0]

for i in range(int(largura/2), int(largura/2 + 30)):
    for j in range(int(altura/2), int(altura/2 + 30)):
        img[i, j] = [0, 0, 255]

plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
# imagem com um quadrado vermelho jeito 2
img_path = 'imagens/Lenna.png'
img = cv2.imread(img_path)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.show()

largura = img.shape[1]
altura = img.shape[0]

img[int(largura/2): int(largura/2 + 30), int(altura/2): int(altura/2 + 30)] = [0, 0, 255]
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
# mostrando cada cor da imagem
img_path = 'imagens/Lenna.png'
img = cv2.imread(img_path)

canalb = img.copy()
canalb[:, :, 1:3] = 0
canalr = img.copy()
canalr[:, :, 0:2] = 0
canalg = img.copy()
canalg[:, :, 0] = 0
canalg[:, :, 2] = 0

plt.imshow(cv2.cvtColor(canalb, cv2.COLOR_BGR2RGB))
plt.show()
plt.imshow(cv2.cvtColor(canalr, cv2.COLOR_BGR2RGB))
plt.show()
plt.imshow(cv2.cvtColor(canalg, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
# converter imagem para tons de cinza
img_path = 'imagens/Lenna.png'
img = cv2.imread(img_path)

gray = img.copy()

largura = img.shape[1]
altura = img.shape[0]

for i in range(largura):
    for j in range(altura):
        gray[i, j] = np.uint8((np.float16(img[i,j,0]) + np.float16(img[i,j,1]) + np.float16(img[i,j,2]))/3)

plt.imshow(gray, cmap='gray')
plt.show()
```

```python
# cortar imagem
img_path = 'imagens/Lenna.png'
img = cv2.imread(img_path)

# sem o copy faz o ROI (regiao de interesse), endereca uma regiao de imagem
new_img = img[250:350, 250:350]
plt.imshow(cv2.cvtColor(new_img, cv2.COLOR_BGR2RGB))
plt.show()
```
